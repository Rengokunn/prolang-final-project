```python
 # Example A.2 - Dynamic Typing
# Python variables can change type at runtime — no declaration needed.
value = 42
print(f"value is {value}, type: {type(value)}")

value = "now I'm a string"
print(f"value is '{value}', type: {type(value)}")

value = [1, 2, 3]
print(f"value is {value}, type: {type(value)}")
```

```python
# Example A.3 - Strong Typing (no silent coercion)
# Python is strongly typed — it won't silently convert types.
num = 5
text = "10"

# This will raise a TypeError:
try:
    result = num + text
except TypeError as e:
    print(f"TypeError caught: {e}")

# Explicit conversion works:
result = num + int(text)
print(f"After explicit conversion: {result}")
```

```python
# Example A.4 - Mutable vs Immutable Reference Types
# Lists are mutable — changes affect all references to the same object.
a = [1, 2, 3]
b = a           # b points to the SAME list object
b.append(4)
print(f"a = {a}")  # [1, 2, 3, 4] — a was also changed!

# Tuples are immutable — they cannot be changed after creation.
coords = (10, 20)
try:
    coords[0] = 99
except TypeError as e:
    print(f"Tuple is immutable: {e}")
```

```python
# Example A.5 - Type Hints (Static-like typing in Python 3.5+)
# Type hints don't enforce types at runtime but help with readability and tools.
def add(x: int, y: int) -> int:
    return x + y

result = add(3, 7)
print(f"add(3, 7) = {result}")

# Python won't stop you passing a wrong type at runtime:
result2 = add("hello", " world")  # no error — hints are not enforced
print(f"add('hello', ' world') = {result2}")
```

```python
# Example A.6 - Collections as Data Types
my_list  = [1, 2, 3, 2]          # ordered, mutable, allows duplicates
my_tuple = (1, 2, 3)             # ordered, immutable
my_set   = {1, 2, 3, 2}         # unordered, unique values only
my_dict  = {"name": "Bob", "age": 30}  # key-value pairs

print(f"list:  {my_list}")
print(f"tuple: {my_tuple}")
print(f"set:   {my_set}")   # duplicate 2 removed
print(f"dict:  {my_dict}")
```

```python
# Example C.3 - while Loop with break and continue
counter = 0
while counter < 10:
    counter += 1

    if counter % 2 == 0:
        continue   # skip even numbers

    if counter == 7:
        break      # stop at 7

    print(f"  counter = {counter}")
```

```python
# Example C.4 - for...else (unique Python feature)
# The else block runs ONLY if the loop was NOT exited by break.
def find_prime(numbers):
    for num in numbers:
        for divisor in range(2, num):
            if num % divisor == 0:
                break          # not prime
        else:
            return num         # loop finished without break = prime found
    return None

nums = [4, 6, 8, 7, 10]
print(f"First prime in {nums}: {find_prime(nums)}")
```

```python
# Example C.5 - Guard Clauses (early return pattern)
# Guards protect the main logic by handling edge cases first.
def calculate_bmi(weight_kg, height_m):
    if weight_kg <= 0:       # guard 1
        return "Error: weight must be positive"
    if height_m <= 0:        # guard 2
        return "Error: height must be positive"
    if height_m > 3:         # guard 3
        return "Error: height seems unrealistic"

    # Main logic only runs if all guards pass
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 2)

print(calculate_bmi(70, 1.75))   # normal
print(calculate_bmi(-5, 1.75))   # guard 1 triggers
print(calculate_bmi(70, 0))      # guard 2 triggers
```

```python
# Example C.6 - Iterating over collections
fruits = ["apple", "banana", "cherry"]

# Basic iteration
for fruit in fruits:
    print(f"  Fruit: {fruit}")

print()
# With index using enumerate()
for index, fruit in enumerate(fruits):
    print(f"  [{index}] {fruit}")

print()
# Iterating two lists together with zip()
prices = [1.20, 0.50, 2.00]
for fruit, price in zip(fruits, prices):
    print(f"  {fruit}: ${price}")
```

```python
# Example C.7 - match / case (Python 3.10+) — Structural Pattern Matching
def describe_input(command):
    match command:
        case "quit" | "exit":
            return "Exiting the program."
        case "help":
            return "Showing help menu."
        case str() if command.startswith("go "):
            direction = command[3:]
            return f"Moving towards: {direction}"
        case _:                        # default case
            return f"Unknown command: '{command}'"

print(describe_input("quit"))
print(describe_input("go north"))
print(describe_input("fly"))
```

```python
# Example D.3 - Recursion
def factorial(n):
    """Calculates n! recursively."""
    if n < 0:
        raise ValueError("n must be non-negative")
    if n == 0 or n == 1:   # base case
        return 1
    return n * factorial(n - 1)   # recursive case

for i in range(7):
    print(f"{i}! = {factorial(i)}")
```

```python
# Example D.4 - Lambda Functions (Anonymous Functions)
square    = lambda x: x ** 2
add       = lambda x, y: x + y
is_even   = lambda x: x % 2 == 0

print(f"square(5) = {square(5)}")
print(f"add(3, 4) = {add(3, 4)}")

# Lambdas are commonly used with map(), filter(), sorted()
numbers = [5, 2, 8, 1, 9, 3]
evens   = list(filter(is_even, numbers))
squared = list(map(square, numbers))
sorted_desc = sorted(numbers, key=lambda x: -x)

print(f"Evens:       {evens}")
print(f"Squared:     {squared}")
print(f"Sorted desc: {sorted_desc}")
```

```python
# Example D.5 - Closures
# A closure is a function that remembers the environment it was created in.
def make_multiplier(factor):
    def multiply(x):          # inner function
        return x * factor     # 'factor' is captured from outer scope
    return multiply           # return the inner function

double = make_multiplier(2)
triple = make_multiplier(3)

print(f"double(5) = {double(5)}")
print(f"triple(5) = {triple(5)}")
print(f"double(10) = {double(10)}")
```

```python
# Example D.6 - Higher-Order Functions and Decorators
import time

# A decorator is a function that wraps another function
def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start  = time.time()
        result = func(*args, **kwargs)
        end    = time.time()
        print(f"  '{func.__name__}' took {(end - start)*1000:.3f} ms")
        return result
    return wrapper

@timer_decorator
def slow_sum(n):
    return sum(range(n))

result = slow_sum(1_000_000)
print(f"  Result: {result}")
```

#### E. Abstract Data Types and Encapsulation
* (Provide 5-10 examples including private/protected members and module exports)

```python
# Example E.1 - Class with Private Attributes (name mangling)
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner           # public
        self._bank = "PyBank"        # protected (convention only)
        self.__balance = balance     # private (name-mangled to _BankAccount__balance)

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit must be positive")
        self.__balance += amount

    def withdraw(self, amount):
        if amount > self.__balance:
            raise ValueError("Insufficient funds")
        self.__balance -= amount

    def get_balance(self):
        return self.__balance

acc = BankAccount("Alice", 100)
acc.deposit(50)
acc.withdraw(30)
print(f"Balance: ${acc.get_balance()}")

# Direct access to private field is blocked:
try:
    print(acc.__balance)
except AttributeError as e:
    print(f"AttributeError: {e}")
```

```python
# Example E.2 - @property Decorator (Controlled Access)
class Temperature:
    def __init__(self, celsius=0):
        self.__celsius = celsius

    @property
    def celsius(self):           # getter
        return self.__celsius

    @celsius.setter
    def celsius(self, value):    # setter with validation
        if value < -273.15:
            raise ValueError("Temperature below absolute zero!")
        self.__celsius = value

    @property
    def fahrenheit(self):        # computed property, no setter
        return self.__celsius * 9/5 + 32

temp = Temperature(25)
print(f"{temp.celsius}°C = {temp.fahrenheit}°F")

temp.celsius = 100
print(f"{temp.celsius}°C = {temp.fahrenheit}°F")

try:
    temp.celsius = -300  # triggers validation
except ValueError as e:
    print(f"ValueError: {e}")
```

```python
# Example E.3 - Abstract Base Class (ABC)
from abc import ABC, abstractmethod

class Shape(ABC):                    # cannot be instantiated directly
    @abstractmethod
    def area(self) -> float:
        pass

    @abstractmethod
    def perimeter(self) -> float:
        pass

    def describe(self):              # concrete method shared by all
        return f"{type(self).__name__}: area={self.area():.2f}, perimeter={self.perimeter():.2f}"

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):      return 3.14159 * self.radius ** 2
    def perimeter(self): return 2 * 3.14159 * self.radius

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w, self.h = w, h
    def area(self):      return self.w * self.h
    def perimeter(self): return 2 * (self.w + self.h)

shapes = [Circle(5), Rectangle(4, 6)]
for s in shapes:
    print(s.describe())

try:
    bad = Shape()   # can't instantiate abstract class
except TypeError as e:
    print(f"TypeError: {e}")
```

```python
# Example E.5 - dataclass (Modern Encapsulation)
from dataclasses import dataclass, field

@dataclass
class Student:
    name:   str
    age:    int
    grades: list = field(default_factory=list)

    def average(self):
        return sum(self.grades) / len(self.grades) if self.grades else 0

    def __str__(self):
        return f"Student({self.name}, age={self.age}, avg={self.average():.1f})"

s = Student("Alice", 20, [88, 92, 79, 95])
print(s)
```

#### F. Object-Oriented Programming
* (Provide 5-10 examples including inheritance, interfaces, and polymorphism)

```python
# Example F.1 - Class, Constructor, and Instance Methods
class Animal:
    species_count = 0             # class variable (shared by all instances)

    def __init__(self, name, sound):
        self.name  = name         # instance variable
        self.sound = sound
        Animal.species_count += 1

    def speak(self):
        return f"{self.name} says '{self.sound}'"

    @classmethod
    def get_count(cls):
        return f"Total animals created: {cls.species_count}"

    def __str__(self):
        return f"Animal(name={self.name})"

cat = Animal("Cat", "Meow")
dog = Animal("Dog", "Woof")

print(cat.speak())
print(dog.speak())
print(Animal.get_count())
```

```python
# Example F.2 - Inheritance
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."

class Dog(Animal):             # Dog inherits from Animal
    def speak(self):           # overrides parent method
        return f"{self.name} barks: Woof!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} meows: Meow!"

class GoldenRetriever(Dog):    # multi-level inheritance
    def fetch(self):
        return f"{self.name} fetches the ball!"

buddy = GoldenRetriever("Buddy")
print(buddy.speak())           # uses Dog's speak
print(buddy.fetch())
print(isinstance(buddy, Dog))    # True
print(isinstance(buddy, Animal)) # True
```

```python
# Example F.3 - Polymorphism
class Animal:
    def __init__(self, name): self.name = name
    def speak(self): return "..."

class Dog(Animal):
    def speak(self): return f"{self.name}: Woof!"

class Cat(Animal):
    def speak(self): return f"{self.name}: Meow!"

class Duck(Animal):
    def speak(self): return f"{self.name}: Quack!"

# Polymorphism: same interface, different behavior
animals = [Dog("Rex"), Cat("Whiskers"), Duck("Donald"), Dog("Buddy")]

for animal in animals:
    print(animal.speak())   # calls each class's own speak()
```

```python
# Example F.4 - Multiple Inheritance and MRO (Method Resolution Order)
class Flyable:
    def move(self): return "Flying through the air"

class Swimmable:
    def move(self): return "Swimming through the water"

class Duck(Flyable, Swimmable):  # inherits from both
    def __init__(self, name):
        self.name = name

donald = Duck("Donald")
print(donald.move())  # Flyable wins — Python uses MRO (left to right)

# Check the resolution order:
print(f"MRO: {[c.__name__ for c in Duck.__mro__]}")
```

```python
# Example F.5 - Dunder (Magic) Methods
class Vector:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __str__(self):           # str() and print()
        return f"Vector({self.x}, {self.y})"

    def __repr__(self):          # developer representation
        return f"Vector(x={self.x}, y={self.y})"

    def __add__(self, other):    # + operator
        return Vector(self.x + other.x, self.y + other.y)

    def __len__(self):           # len()
        return int((self.x**2 + self.y**2) ** 0.5)

    def __eq__(self, other):     # == operator
        return self.x == other.x and self.y == other.y

v1 = Vector(3, 4)
v2 = Vector(1, 2)
v3 = v1 + v2

print(v1)
print(f"v1 + v2 = {v3}")
print(f"len(v1) = {len(v1)}")
print(f"v1 == v2: {v1 == v2}")
print(f"v1 == Vector(3,4): {v1 == Vector(3, 4)}")
```

```python
# Example G.1 - Threading (I/O-bound tasks)
import threading
import time

def download_file(filename, duration):
    print(f"  Starting download: {filename}")
    time.sleep(duration)          # simulate download
    print(f"  Finished download: {filename}")

start = time.time()

# Create and start two threads
t1 = threading.Thread(target=download_file, args=("file_A.zip", 2))
t2 = threading.Thread(target=download_file, args=("file_B.zip", 1))

t1.start()
t2.start()

t1.join()   # wait for t1 to finish
t2.join()   # wait for t2 to finish

elapsed = time.time() - start
print(f"Both files downloaded in {elapsed:.2f}s (ran concurrently!)")
```

```python
# Example G.2 - asyncio and async/await
import asyncio

async def fetch_data(source, delay):
    print(f"  Fetching from {source}...")
    await asyncio.sleep(delay)    # non-blocking wait
    return f"Data from {source}"

async def main():
    start = asyncio.get_event_loop().time()

    # Run both coroutines concurrently
    results = await asyncio.gather(
        fetch_data("API_1", 2),
        fetch_data("API_2", 1),
        fetch_data("API_3", 1.5)
    )

    for r in results:
        print(f"  Result: {r}")

# In Jupyter Notebook, use:
await main()

# In a regular .py script, use instead:
# asyncio.run(main())
```

```python
# Example G.3 - Thread Synchronization (Lock)
import threading

counter = 0
lock    = threading.Lock()

def increment(n):
    global counter
    for _ in range(n):
        with lock:             # acquire lock, auto-release after block
            counter += 1

threads = [threading.Thread(target=increment, args=(1000,)) for _ in range(5)]
for t in threads: t.start()
for t in threads: t.join()

print(f"Final counter: {counter}")  # should be 5000 (no race condition)
```

```python
# Example G.4 - concurrent.futures.ThreadPoolExecutor
from concurrent.futures import ThreadPoolExecutor
import time

def process_item(item):
    time.sleep(0.1)          # simulate work
    return item * item

items = list(range(10))

start = time.time()
with ThreadPoolExecutor(max_workers=4) as executor:
    results = list(executor.map(process_item, items))
elapsed = time.time() - start

print(f"Results:  {results}")
print(f"Elapsed:  {elapsed:.2f}s")
```

```python
# Example G.5 - multiprocessing (CPU-bound, bypasses the GIL)
from multiprocessing import Pool
import time

def cpu_heavy(n):
    """Compute sum of squares — CPU-bound."""
    return sum(i*i for i in range(n))

data = [1_000_000] * 4

# Sequential
start = time.time()
seq_results = [cpu_heavy(n) for n in data]
seq_time = time.time() - start

# NOTE: On Windows, multiprocessing requires the code to be inside
# an 'if __name__ == "__main__":' guard when run as a .py script.
# This cell runs correctly inside Jupyter Notebook without the guard.

# Parallel with multiprocessing
start = time.time()
with Pool(processes=4) as pool:
    par_results = pool.map(cpu_heavy, data)
par_time = time.time() - start

print(f"Sequential time: {seq_time:.3f}s")
print(f"Parallel time:   {par_time:.3f}s")
print(f"Results match:   {seq_results == par_results}")
```

#### H. Exception and Event Handling
* (Provide 5-10 examples including try/catch, custom exceptions, or event listeners)

```python
# Example H.1 - Basic try/except/else/finally
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("  Caught: Cannot divide by zero!")
        return None
    except TypeError as e:
        print(f"  Caught TypeError: {e}")
        return None
    else:
        print(f"  Success: {a} / {b} = {result}")  # runs only if no exception
        return result
    finally:
        print("  Finally block always runs.")        # always runs

safe_divide(10, 2)
print()
safe_divide(10, 0)
print()
safe_divide(10, "a")
```

```python
# Example H.2 - Catching Multiple Exception Types
data = ["10", "0", "abc", None, "5"]

for item in data:
    try:
        result = 100 / int(item)
        print(f"  100 / int('{item}') = {result:.2f}")
    except ZeroDivisionError:
        print(f"  '{item}' → ZeroDivisionError")
    except ValueError:
        print(f"  '{item}' → ValueError (can't convert to int)")
    except TypeError:
        print(f"  '{item}' → TypeError (item is None)")
```

```python
# Example H.3 - Custom Exception Classes
class AppError(Exception):           # base custom exception
    pass

class InsufficientFundsError(AppError):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount  = amount
        super().__init__(f"Need ${amount - balance:.2f} more (have ${balance:.2f}, need ${amount:.2f})")

class AccountLockedError(AppError):
    def __init__(self, account_id):
        super().__init__(f"Account {account_id} is locked.")

def withdraw(balance, amount, locked=False):
    if locked:
        raise AccountLockedError("ACC-001")
    if amount > balance:
        raise InsufficientFundsError(balance, amount)
    return balance - amount

for test in [(100, 50, False), (100, 200, False), (100, 50, True)]:
    try:
        new_balance = withdraw(*test)
        print(f"  Withdrew ${test[1]}, new balance: ${new_balance}")
    except AppError as e:   # catches any AppError subclass
        print(f"  {type(e).__name__}: {e}")
```

```python
# Example H.5 - raise, re-raise, and Exception Chaining
def parse_age(value):
    try:
        age = int(value)
    except ValueError as e:
        # Chain the original error for context
        raise ValueError(f"Invalid age input: '{value}'") from e

    if age < 0 or age > 150:
        raise ValueError(f"Age {age} is out of realistic range (0–150)")

    return age

test_inputs = ["25", "abc", "-5", "200", "42"]

for val in test_inputs:
    try:
        age = parse_age(val)
        print(f"  '{val}' → valid age: {age}")
    except ValueError as e:
        print(f"  '{val}' → ValueError: {e}")
```

```python
# Example H.6 - Custom Context Manager using __enter__ / __exit__
class ManagedResource:
    def __init__(self, name):
        self.name = name

    def __enter__(self):
        print(f"  [{self.name}] Resource acquired.")
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        print(f"  [{self.name}] Resource released.")
        if exc_type:
            print(f"  [{self.name}] Exception was: {exc_val}")
        return False   # don't suppress exceptions

    def do_work(self):
        print(f"  [{self.name}] Doing work...")

with ManagedResource("DB Connection") as res:
    res.do_work()

print()

try:
    with ManagedResource("File Handle") as res:
        res.do_work()
        raise RuntimeError("Simulated error mid-operation")
except RuntimeError as e:
    print(f"  Caught outside: {e}")
```

## 4. "Code Smell" & Refactoring Reflection

**Original Code Link:** https://github.com/Rengokunn/prolang-final-project

**The "Smell":** The function uses a mutable list as a default parameter value. In Python, default parameter values are evaluated once when the function is defined, not each time it is called. This means the same list object is reused across all calls, causing data to accumulate silently. This violates the Principle of Least Surprise — the function appears to create a fresh cart each time, but it does not. This is a well-known Python gotcha that stems from how objects are stored in memory at definition time.

**The Refactor:** The fix is to use None as the default parameter and initialize a fresh list inside the function body, ensuring each call gets its own independent list.

```python
# SMELLY CODE — Mutable default argument
def add_item_bad(item, cart=[]):   # 'cart' is shared across ALL calls!
    cart.append(item)
    return cart

print(add_item_bad("apple"))    # ['apple'] — looks fine
print(add_item_bad("banana"))   # ['apple', 'banana'] — BUG: 'apple' is still there!
print(add_item_bad("cherry"))   # ['apple', 'banana', 'cherry'] — keeps accumulating
```

```python
# CLEAN REFACTOR — Use None as default sentinel
def add_item_clean(item, cart=None):
    if cart is None:
        cart = []                  # create a fresh list each call
    cart.append(item)
    return cart

print(add_item_clean("apple"))    # ['apple']
print(add_item_clean("banana"))   # ['banana'] — fresh list every time!

# You can still pass an existing cart:
my_cart = ["existing_item"]
print(add_item_clean("cherry", my_cart))  # ['existing_item', 'cherry']
```
